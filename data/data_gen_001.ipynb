```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook
```

```python
num = 50

x1 = np.linspace(0,4*np.pi,num)
x2 = np.linspace(0,4*np.pi,num)

xx1,xx2 = np.meshgrid(x1,x2)

y1 = np.sin(xx1)+np.cos(xx2)+0.2*(np.random.rand(num,num)-0.5)

data_input = np.array([xx1, xx2]).reshape(2, -1).T
data_output = np.array([y1]).reshape(1, -1).T

df_input = pd.DataFrame(data_input, columns=['x1','x2'])
df_output = pd.DataFrame(data_output, columns=['y1'])

df_input.to_csv('data_input_001.csv')
df_output.to_csv('data_output_001.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xx1,xx2,y1,
           edgecolor = 'black',
           alpha = 0.3)
```
