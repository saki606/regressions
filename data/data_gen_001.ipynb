```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook
```

```python
num_x1 = 50
num_x2 = 50
num_x3 = 50

# x1 = np.linspace(0,4*np.pi,num_x1)
# x2 = np.linspace(0,4*np.pi,num_x2)

x1 = 4*np.pi*(np.random.uniform(size=num_x1)-0.5)
x2 = 4*np.pi*(np.random.uniform(size=num_x2)-0.5)
x3 = (np.random.uniform(size=num_x3)-0.5)

xx1,xx2 = np.meshgrid(x1,x2)

y1 = np.sin(xx1)+np.cos(xx2)+0.2*(np.random.rand(num_x1,num_x2)-0.5)

data_input = np.array([xx1, xx2]).reshape(2, -1).T
data_output = np.array([y1]).reshape(1, -1).T

df_input = pd.DataFrame(data_input, columns=['x1','x2'])
df_output = pd.DataFrame(data_output, columns=['y1'])

df_input.to_csv('data_input_001.csv')
df_output.to_csv('data_output_001.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xx1,xx2,y1,
           edgecolor = 'black',
           alpha = 0.3)
```
