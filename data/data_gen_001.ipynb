```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook
```

```python
num = 50

x = np.linspace(0,4*np.pi,num)
y = np.linspace(0,4*np.pi,num)

xx,yy = np.meshgrid(x,y)

z = np.sin(xx)+np.cos(yy)+0.2*(np.random.rand(num,num)-0.5)

data = np.array([xx, yy, z]).reshape(3, -1).T
df = pd.DataFrame(data, columns=['x','y','z'])

df.to_csv('data_001.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xx,yy,z,
           edgecolor = 'black',
           alpha = 0.3)
```
